```python
import tensorflow  
import numpy as np
from numpy import linalg
from numpy.linalg import norm
from tensorflow import keras 
import os 
from tqdm import tqdm
from tensorflow.keras.preprocessing import image 
from tensorflow.keras.layers import GlobalMaxPool2D
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input

model = ResNet50(weights= "imagenet", include_top= False, input_shape=(224, 224, 3))
model.trainable = False

model = tensorflow.keras.Sequential([

    model, 
    GlobalMaxPool2D()

])
#print(model.summary())
```

```python
def extract_features(image_path , model):
    img = image.load_img(image_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    exp_img_array = np.expand_dims(img_array, axis = 0)
    preprocessed_img = preprocess_input(exp_img_array)
    results = model.predict(preprocessed_img).flatten()
    normalized_results = results/norm(results)
    return normalized_results

print

filename = []
for file in os.listdir("images"):
    filename.append(os.path.join("images", file))

#print(os.listdir("images"))

features_list = []
for file_path in tqdm(filename):
    features_list.append(extract_features(file_path, model))
features_list = np.array(features_list)
print(features_list)





```
